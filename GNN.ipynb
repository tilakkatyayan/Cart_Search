# MultiEdge Attribute
Create adjacency matrices for multiple edge attributes with structured dtypes:

```python
import numpy as np

G = nx.Graph()
G.add_edge(0, 1, weight=10)
G.add_edge(1, 2, cost=5)
G.add_edge(2, 3, weight=3, cost=-4.0)
dtype = np.dtype([("weight", int), ("cost", float)])
A = nx.to_numpy_array(G, dtype=dtype, weight=None)
print("weight --------")
print(A["weight"])   # creating adjacency matrix of weight
print("cost --------")
print(A["cost"])     # creating adjacency matrix of cost
nx.draw(G)
print(G.edges)
print(G.nodes)
```

Returning adjacency representation of graph as a dictionary of dictionaries.

```python
G = nx.Graph()
G.add_edges_from(
    [
        (0, 1, {'weight': 1.0}),
        (1, 2, {'weight': 2.0}),
        (2, 0, {'weight': 1.0}),
    ]
)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
d = nx.to_dict_of_dicts(G) # represent in dictionary of dictionary
print(d)
print(d[1][2]['weight'])
```

# Multigraph
It contains multiple edges but not directional

```python
G = nx.MultiGraph()
G.add_edge(0, 1, key='a', weight=1.0)
G.add_edge(0, 1, key='b', weight=5.0)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
d = nx.to_dict_of_dicts(G)
print(d)
print(d[0][1]['b']['weight'])
```

Splitting Graph into a pandas dataframe

```python
G = nx.Graph(
    [
        ("A", "B", {"cost": 1, "weight": 7}),
        ("C", "E", {"cost": 9, "weight": 10}),
    ]
)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
df = nx.to_pandas_edgelist(G)  # it gives pandas dataframe according to main graph
print(df)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=["B", "E"])  # it gives pandas dataframe according to node list B and E means source is B and E
print(df)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=["A", "C"])  # it gives pandas dataframe according to node list B and E means source is A and C
print(df)
print("--------------------")
df[["source", "target", "cost", "weight"]]    # printing value of each
```

Printing Adjacency Matrix

```python
G = nx.Graph(
    [
        ("A", "B", {"cost": 1, "weight": 7}),
        ("C", "E", {"cost": 9, "weight": 10}),
    ]
)
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
df = nx.to_pandas_edgelist(G)
print(df)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=["B", "E"])
print(df)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=["A", "C"])
print(df)
print("--------------------")
df[["source", "target", "cost", "weight"]]
print("--------------------")
A = nx.adjacency_matrix(G)  # by default printing adjacency matrix of A with weight
print(A.todense())
```

```python
G = nx.MultiGraph([('A', 'B', {'cost': 1}), ('A', 'B', {'cost': 9})])
nx.draw(G)
print(G.edges)
print(G.nodes)
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=['A', 'C'], edge_key='ekey')
print(df)
print(df[['source', 'target', 'cost', 'ekey']])
print("--------------------")
df = nx.to_pandas_edgelist(G, nodelist=['B'], edge_key='ekey')
print(df)
print(df[['source', 'target', 'cost', 'ekey']])
```
